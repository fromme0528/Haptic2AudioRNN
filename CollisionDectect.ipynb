# Audio to Spectrogram

```python
import hparams as hp
from hparams import Default as hp_default
import pickle

def wav2spctro(filePath):
    result = list()
    #load audio
    audio, rate = librosa.load(filePath, mono=True, sr = hp_default.sr)
    print('audio :', audio.shape[0],'rate :', rate)
    spectro = librosa.stft(audio,
        n_fft = hp_default.n_fft,
        hop_length = hp_default.hop_length,
        win_length = hp_default.win_length
        )
    print ('spectro',spectro.shape[0],spectro.shape[1])
    #Normaliztio

    #paddding
    xsize = int(1+hp_default.sr*hp_default.timeLength/hp_default.hop_length)
    if spectro.shape[1] < xsize:
        print('add padding')
        spectro = np.lib.pad(spectro, ((0, 0), (0, xsize - spectro.shape[1])), 'constant', constant_values = 1e-13)
        spectro_ = spectro.T
    if spectro.shape[1] > xsize:
        spectro_ = spectro.T[:xsize][:]
        print('cut')
    if (spectro.shape[1] == xsize):
        spectro_ = spectro.T

    if spectro_.shape[0] == xsize:
        result.append(np.abs(spectro_))

    return result[0]

outPath = "./dataset/spectrogram"
inPath = "./dataset/audio/"

if not os.path.exists(outPath):
    os.makedirs(outPath)
    
for path, dirs, files in os.walk(inPath):
    for f in files:
        if os.path.splitext(f)[-1] == '.wav':
            print(os.path.join(path,f))
            spectro = wav2spctro(inPath+f)
            outFile = 'spectro' + os.path.splitext(f)[0][5:] + '.pickle'
            with open(os.path.join(outPath,outFile),'wb') as fs:
                pickle.dump(spectro,fs)
                
```

# Collision Detection for Accel

## ./dataset/accel_plastic.csv

```python
import os
import csv
```

```python

sr_accel = 500
accelList = list()
temp_accel = 0

with open(os.path.join("./dataset/accel_plastic.csv"),'r',encoding='UTF8') as csvfile:
    
    data_accel = csv.reader(csvfile)
    data_accel = [line for line in data_accel]
    
    for idx, data in enumerate(data_accel):
        
        if idx < temp_accel + 2*sr_accel/3:
            continue
            
        if np.abs(float(data[0])) + np.abs(float(data[1])) + np.abs(float(data[2])) > 20 :
            a1 = idx - int(sr_accel/5)
            a2 = idx + int(2*sr_accel/3)
            collision_idx = np.argmax(data_accel[a1:a2][2])
            accelList.append((collision_idx+idx))
            temp_accel = idx
```

```python
print(len(accelList))
print(len(data_accel)/sr_accel)
```

```python
inPath_folder = "./dataset/accel_250/"

files_accel = os.listdir(inPath_folder) 
files_accel = [f for f in files_accel if os.path.splitext(f)[-1] == '.csv']

for inPath in files_accel:
# 'cp949' codec can't decode
    print(inPath_folder + inPath)
    with open(os.path.join(inPath_folder + inPath),'r',encoding='UTF8') as csvfile:

        data_accel = csv.reader(csvfile)
        data_accel = [line for line in data_accel]

        output = list()

        for idx,line in enumerate(data_accel) :

            line = [float(i) for i in line]

            if idx != (len(data_accel) - 1):
                line_next = [float(i) for i in data_accel[idx+1]]
#                    interval = list(set(data_accel[idx+1][0:2]) - set(line[0:2]))
                interval = list()
                interval.append(line_next[0] - line[0])
                interval.append(line_next[1] - line[1])
                interval.append(line_next[2] - line[2])

                for i in range(0,32,1):
                    temp = list()
                    temp.append(line[0] + (interval[0] * i * 0.03125))
                    temp.append(line[1] + (interval[1] * i * 0.03125))
                    temp.append(line[2] + (interval[2] * i * 0.03125))
                    
                    temp.append(line[-1])
                    output.append(temp)
            else:
                for i in range(0,32,1):
                    output.append(line)
    with open (os.path.join("dataset/accel_8000/"+inPath),'w',newline='') as fs:
        wr = csv.writer(fs)
        for row in output:
            wr.writerow(row)
```
