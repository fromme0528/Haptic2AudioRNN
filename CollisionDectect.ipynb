# Collision Detection for Accel

## ./dataset/accel_plastic.csv

```python
import os
import csv
```

```python

sr_accel = 500
accelList = list()
temp_accel = 0

with open(os.path.join("./dataset/accel_plastic.csv"),'r',encoding='UTF8') as csvfile:
    
    data_accel = csv.reader(csvfile)
    data_accel = [line for line in data_accel]
    
    for idx, data in enumerate(data_accel):
        
        if idx < temp_accel + 2*sr_accel/3:
            continue
            
        if np.abs(float(data[0])) + np.abs(float(data[1])) + np.abs(float(data[2])) > 20 :
            a1 = idx - int(sr_accel/5)
            a2 = idx + int(2*sr_accel/3)
            collision_idx = np.argmax(data_accel[a1:a2][2])
            accelList.append((collision_idx+idx))
            temp_accel = idx
```

```python
print(len(accelList))
print(len(data_accel)/sr_accel)
```

```python
print(accelList)
```

```python
with open(os.path.join("./dataset/accel_plastic.csv"),'r',encoding='UTF8') as csvfile:

    data_accel = csv.reader(csvfile)
    data_accel = [line for line in data_accel]
    
    for num, idx in enumerate(accelList):
        result = data_accel[idx-49:idx+201]
        with open (os.path.join("dataset/accel_250/accel_plastic_"+str(num)+'.csv'),'w',newline='') as fs:
            wr = csv.writer(fs)
            for row in result:
                wr.writerow(row)
        result = list()
```
